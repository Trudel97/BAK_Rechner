```python
pdfreader = PdfReader("Warenkauf-Richtlinie-RL_2019-771-EU.pdf")
```

```python
from typing_extensions import Concatenate

raw_text = ""
for i, page in enumerate(pdfreader.pages):
  content = page.extract_text()
  if content:
    raw_text += content
```

```python
cassio.init(token=ASTRA_DB_APPLICATION_TOKEN, database_id=ASTRA_DB_ID)
```

```python
llm = OpenAI(openai_api_key=OPENAI_API_KEY)
embedding= OpenAIEmbeddings(openai_api_key=OPENAI_API_KEY)
```

```python
astra_vector_store = Cassandra(
    embedding=embedding,
    table_name="qa_mini_demo",
    session=None,
    keyspace=None
)
```

```python
from langchain.text_splitter import CharacterTextSplitter

text_splitter = CharacterTextSplitter(
    separator= "\n",
    chunk_size = 800,
    chunk_overlap = 200,
    length_function = len
)

texts = text_splitter.split_text(raw_text)
```

```python
texts[:50]
```

```python
astra_vector_store.add_texts(texts[:50])

print("Inserted %i headlines." % len(texts[:50]))

astra_vector_index = VectorStoreIndexWrapper(vectorstore=astra_vector_store)
```

```python
first_question = True
while True:
    if first_question:
        query_text = input("\nEnter your first question (or type 'quit' to exit): ").strip()
    else:
        query_text = input("\nWhat's your next question (or type 'quit' to exit): ").strip()

    if query_text.lower() == "quit":
        break
    if query_text == "":
        continue

    first_question = False

    print('\nQUESTION: "%s"' % query_text)
    # Annahme: astra_vector_index.query und die folgenden Methoden sind korrekt definiert
    answer = astra_vector_index.query(query_text, llm=llm).strip()
    print('ANSWER: "%s"\n' % answer)

    print("FIRST DOCUMENTS BY RELEVANCE:")
    for doc, score in astra_vector_store.similarity_search_with_score(query_text, k=4):
        print("   [%0.4f] \"%s ...\"" % (score, doc.page_content[:84]))
```
